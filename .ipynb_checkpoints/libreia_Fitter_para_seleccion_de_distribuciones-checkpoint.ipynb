## Encontrar la mejor distribución de los datos en Python:

Con Fitter, podemos ajustar los datos a varias distribuciones comunes y obtener los parámetros de cada distribución. 

Luego, podemos comparar la calidad del ajuste de cada distribución utilizando diferentes métricas, como la raíz del error cuadrático medio o el coeficiente de determinación. Finalmente, podemos seleccionar la distribución que mejor se ajuste a tus datos.

A continuación, aplicamos la librería con ejemplo de cómo se utiliza Fitter para ajustar los datos a diferentes distribuciones y seleccionar la mejor distribución:

```python
!pip install fitter
```

```python
import numpy as np
from fitter import Fitter
```

```python
# Generar algunos datos de ejemplo
data = np.random.normal(0, 1, size=1000)
```

```python
data
```

```python
# Crear un objeto Fitter y ajustar los datos a varias distribuciones
f = Fitter(data)
f.fit()
```

```python
# Obtener una lista de las distribuciones probadas y sus parámetros
print(f.summary())
```

```python
# Seleccionar la mejor distribución según el criterio BIC, se intentó con AIC pero no lo reconoció
print(f.get_best(method='bic'))
```

Podemos observar que el mejor metodo, para la base de datos ficticia que creamos (Con distribución normal), efectivamente nos dio la distribución normal como la mejor distribución.
